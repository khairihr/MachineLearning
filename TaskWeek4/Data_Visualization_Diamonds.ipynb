```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Ganti path sesuai dengan path ke dataset Anda di Google Drive
file_path = '/content/drive/MyDrive/Dataset MachineLearning/diamonds.csv'

# Membaca dataset menjadi DataFrame
df = pd.read_csv(file_path)

# Sekarang 'df' adalah DataFrame yang berisi data dari dataset Anda
df.head()
```

```python
features = ['carat','cut','color','clarity','depth','table','price','y','x','z']

for feature in features:
  print(f"Jumlah unique value for{feature}: {df[feature].nunique()}")
```

```python
# Kelompokkan fitur ke dalam kategori
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']
categorical_features = ['cut', 'color', 'clarity']
target = ['price']

# Print kategori fitur
print("Numerical Features:")
print(numerical_features)

print("\nCategorical Features:")
print(categorical_features)

print("\nTarget:")
print(target)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengambil daftar fitur kategoris
categorical_features = ['cut', 'color', 'clarity']

# Menghitung jumlah kolom
num_features = len(categorical_features)

# Menghasilkan warna secara dinamis berdasarkan jumlah kolom
colors = sns.color_palette("Set2", n_colors=num_features)

for i, feature in enumerate(categorical_features):
    feature_counts = df[feature].value_counts()
    plt.figure(figsize=(8, 6))
    feature_counts.plot(kind='bar', color=colors[i])
    plt.title(f"Distribusi {feature}")
    plt.xlabel(feature)
    plt.ylabel("Jumlah")
    plt.xticks(rotation=0)
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Daftar fitur kategoris
categorical_features = ['cut', 'color', 'clarity']

for feature in categorical_features:
    feature_counts = df[feature].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f"Distribusi {feature}")
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Daftar fitur numerik
numerical_features = ["carat", "depth", "table", "x", "y", "z"]

# Mengatur ukuran gambar (figsize) lebih besar
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    ax = axes[row, col]

    ax.hist(df[feature], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f"Distribusi {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Jumlah")

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Daftar fitur numerik
numerical_features = ["carat", "depth", "table", "x", "y", "z"]

# Membuat histogram interaktif dengan lebar bin yang dapat disesuaikan
for feature in numerical_features:
    fig = px.histogram(df, x=feature, nbins=20, title=f"Distribusi {feature}")
    fig.update_traces(marker_color='skyblue', marker_line_width=1, marker_line_color='black')
    fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Daftar fitur numerik
numerical_features = ["carat", "depth", "table", "x", "y", "z"]

# Membuat subplot dengan tata letak 2 baris dan 3 kolom
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Membuat boxplot horizontal untuk setiap fitur
for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    ax = axes[row, col]

    sns.boxplot(data=df, y=feature, orient="h", ax=ax)
    ax.set_title(f"Distribusi {feature}")
    ax.set_ylabel(feature)
    ax.set_xlabel("")

plt.tight_layout()
plt.show()
```
