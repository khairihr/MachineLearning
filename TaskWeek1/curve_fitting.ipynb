```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#mendefinisikan fungsi yang akan dilakukan fitting (contoh: fungsi kuadrat sederhana)
def fungsi_kuadrat(x, a, b, c):
    return a * x**2 + b * x + c
```

```python
#contoh data angka random/bebas
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 3.9, 6.8, 9.8, 14.2])
```

```python
#curve fitting
parameter, kovariansi = curve_fit(fungsi_kuadrat, x_data, y_data)

a, b, c = parameter
```

```python
#membuat kurva dari parameter-parameter yang ada
x_kurva = np.linspace(0, 6, 100)
y_kurva = fungsi_kuadrat(x_kurva, a, b, c)
```

```python
plt.scatter(x_data, y_data, label='Titik Data')
plt.plot(x_kurva, y_kurva, color='black', label='Kurva yang dilakukan fitting')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
print(f'Parameternya (a, b, c): {a}, {b}, {c}')
```
