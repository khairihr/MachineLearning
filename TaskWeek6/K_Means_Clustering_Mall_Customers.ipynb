```python
import pandas as pd

# Ganti 'file.csv' dengan nama file dataset yang ingin Anda baca
file_path = '/content/drive/MyDrive/Dataset MachineLearning/Mall_Customers.csv'

# Membaca dataset CSV menjadi DataFrame
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari DataFrame
df.head()
```

```python
# Mengubah nama atribut/kolom
df.rename(columns={'Annual Income (k$)': 'Annual Income', 'Spending Score (1-100)': 'Spending Score'}, inplace=True)

# Menampilkan DataFrame setelah mengubah nama atribut
df.head()
```

```python
# Menghapus kolom 'CustomerID' dan 'Gender'
df = df.drop(['CustomerID', 'Gender'], axis=1)
df.head()
```

Beri kode untuk memvisualisasikan elbow method dengan maksimal jumlah k=10 pada dataframe bernama df, dan nama columns 'Age', 'Annual Income', 'Spending Score'

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = df[['Age', 'Annual Income', 'Spending Score']]

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

    plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

dengan dataframe bernama df, beri kode untuk membuat model k means clustering yang berisi atribut 'Age', 'Annual Income', 'Spending Score'; jumlah cluster=5; dan beri kolom baru bernama label dan visualisasikan berdasarkan label.

```python
# Memilih atribut yang akan digunakan
data = df[['Age', 'Annual Income', 'Spending Score']]

# Membuat model K-Means dengan 5 kluster
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(data)

# Menambahkan kolom "label" ke DataFrame
df['label'] = kmeans.labels_

import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income', y='Spending Score', data=df, hue='label', palette='Set1', s=100)
plt.title('K-Means Clustering (k=5)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(title='Cluster', loc='upper right')
plt.show()
```

Beri kode untuk menghitung nilai sillhouette score untuk k=2 hingga k=10

```python
from sklearn.metrics import silhouette_score
import numpy as np

data = df[['Age', 'Annual Income', 'Spending Score']]

# Rentang nilai k yang akan diuji
k_range = range(2, 11)

silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(data)
    silhouette_avg = silhouette_score(data, labels)
    silhouette_scores.append(silhouette_avg)

# Plot nilai Silhouette Score untuk setiap nilai k
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

Beri kode untuk menghitung nilai Davies Bouldin Index score untuk k=2 hingga k=10

```python
from sklearn.metrics import davies_bouldin_score

data = df[['Age', 'Annual Income', 'Spending Score']]

# Rentang nilai k yang akan diuji
k_range = range(2, 11)

davies_bouldin_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(data)
    davies_bouldin = davies_bouldin_score(data, labels)
    davies_bouldin_scores.append(davies_bouldin)

# Plot nilai Davies-Bouldin Index Score untuk setiap nilai k
plt.figure(figsize=(8, 6))
plt.plot(k_range, davies_bouldin_scores, marker='o', linestyle='-')
plt.title('Davies-Bouldin Index Score for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Davies-Bouldin Index Score')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score

data = df[['Age', 'Annual Income', 'Spending Score']]

# Rentang nilai k yang akan diuji
k_range = range(2, 11)

calinski_harabasz_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(data)
    calinski_harabasz = calinski_harabasz_score(data, labels)
    calinski_harabasz_scores.append(calinski_harabasz)

# Plot nilai Calinski-Harabasz Index Score untuk setiap nilai k
plt.figure(figsize=(8, 6))
plt.plot(k_range, calinski_harabasz_scores, marker='o', linestyle='-')
plt.title('Calinski-Harabasz Index Score for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Calinski-Harabasz Index Score')
plt.grid(True)
plt.show()
```
